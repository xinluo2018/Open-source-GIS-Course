### 介绍开源制图库cartopy绘图基础
Cartopy是一个Python包，用于地理空间数据处理，以便生成地图和其他地理空间数据分析。Cartopy最初为英国气象局开发的，目的是让科学家能够快速、方便以及准确地在地图上可视化他们的数据。Cartopy在开源协议GNU Lesser下对用户开放。
Cartopy是在matplotlib库基础上开发的，其与matplotlib的区别在于添加了与丛中与地理信息相关的功能，如添加投影、添加海岸线等。



```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeat
import numpy as np
from osgeo import ogr
import matplotlib
```

#### cartopy投影：等距圆柱投影(platecarre, cartopy默认投影), 阿尔伯斯等面积投影(AlbersEqualArea), 等角正切圆柱投影(Mercator)
参考：https://scitools.org.uk/cartopy/docs/latest/reference/projections.html


```python
#### 添加地理要素、添加经纬度及格网
proj = ccrs.PlateCarree()
x_extent = [-180,-120,-60,0,60,120,180]  ## 经度范围
y_extent = [-90, -60, -30, 0, 30, 60,90]    ## 纬度范围
fig = plt.figure(figsize=(6,6))
ax_geo = fig.add_subplot(1,1,1, projection=proj)
ax_geo.add_feature(cfeat.LAND)
ax_geo.add_feature(cfeat.RIVERS)
ax_geo.coastlines()
ax_geo.set_xticks(x_extent)    ## 添加经纬度
ax_geo.set_yticks(y_extent)
ax_geo.gridlines(linestyle='--', color='grey')
```

```python
proj_1 = ccrs.PlateCarree()
proj_2 = ccrs.AlbersEqualArea()
proj_3 = ccrs.Mercator()
fig = plt.figure(figsize=(10,3))
# ax = fig.add_subplot(1,1,1)
ax_geo_1 = fig.add_subplot(1,3,1, projection=proj_1)
ax_geo_1.coastlines()
ax_geo_1.add_feature(cfeat.LAND)
ax_geo_1.set_title('distance-equal projection')
ax_geo_2 = fig.add_subplot(1,3,2, projection=proj_2)
ax_geo_2.coastlines()
ax_geo_2.add_feature(cfeat.LAND)
ax_geo_2.set_title('area-equal projction')
ax_geo_3 = fig.add_subplot(1,3,3, projection=proj_3)
ax_geo_3.coastlines()
ax_geo_3.add_feature(cfeat.LAND)
ax_geo_3.set_title('angle-equal projection')
```

```python
fig = plt.figure(figsize=(6,6))
ax_geo_1 = fig.add_subplot(1,1,1, projection=proj_1)
ax_geo_1.add_feature(cfeat.LAND)
ax_geo_1.add_feature(cfeat.RIVERS, edgecolor='red')
ax_geo_1.set_extent([60.0, 120.0, 10, 50.0])
x_extent = [60.0, 70, 80, 90, 100, 110, 120.0]## 经度范围
y_extent = [10, 20, 30, 40, 50]    ## 纬度范围
ax_geo_1.set_xticks(x_extent)    ## 添加经纬度
ax_geo_1.set_yticks(y_extent)
```
